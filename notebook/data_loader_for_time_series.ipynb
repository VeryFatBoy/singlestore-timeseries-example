```python
import pandas as pd
```

```python
tick_csv_url = "https://raw.githubusercontent.com/VeryFatBoy/singlestore-timeseries-example/refs/heads/main/dataset/all_stocks_5yr.csv"

tick_df = pd.read_csv(tick_csv_url)
```

```python
tick_df = tick_df.dropna()
```

```python
tick_df.count()
```

```python
tick_df.drop(columns = ["open", "high", "low", "volume"], inplace = True)
tick_df.rename(columns = {"date": "ts", "close": "price", "Name": "symbol"}, inplace = True)
tick_df = tick_df.sort_values(by = ["ts", "symbol"])
```

```python
from sqlalchemy import *

db_connection = create_engine(connection_url)
```

```python
tick_df.to_sql(
    "tick",
    con = db_connection,
    if_exists = "append",
    index = False,
    chunksize = 1000
)
```
